```python
#Lister les aliments dont le prix est compris entre 100 et 200, et qui sont des « viandes »
df.loc[ (df["Prix"]>100) & (df["Prix"]<200) & (df["Categorie"]== "Viandes") ][['Nom','Categorie']]
```

```python
#Lister les 15 produits les moins chers
df.nsmallest(15, "Prix")
  
```

```python
#Calculer la moyenne de prix des boissons distribuées à Lyon (pivot_table)
tab1 =df.loc[(df['Ville']=='Lyon')&(df["Categorie"]=="Boissons"),:]                  

tab1.pivot_table(index=['Categorie'],columns=['Ville'],values=['Prix'],aggfunc=pd.Series.mean)
```

```python
#Quels sont les 5 produits les moins chers vendus à Lyon 
dta= df.loc[df["Ville"]=="Lyon",:]
dta.nsmallest(5,"Prix")
```

```python
#En utilisant une boucle afficher la representation graphique de toutes les variables quantitative
list_qual=[]
list_quant=[]
for index in df.columns:
    if df[index].dtypes=='O':
        list_qual.append(index)
    else:
        list_quant.append(index)

list_quant
plt.figure(figsize=[17,6])

plt.gcf().subplots_adjust( hspace = 0.5) 
i=1
for var in list_quant :
    plt.subplot(2,4,i)
    plt.hist(df[var],edgecolor='white')
    plt.title(var)
    i+=1
```
